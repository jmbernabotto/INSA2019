```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier

# importons une base de données de chiffre
digits = load_digits()
X = digits.data
y = digits.target
print('dimension de X:', X.shape)

# visualisons un de ces chiffres
plt.figure()
plt.imshow(digits['images'][0], cmap = 'Greys_r')
plt.show()
# Entraînement du modele
model = KNeighborsClassifier()
model.fit(X, y)
model.score(X, y) #ici il s'agit de la moyenne des mesures de l'accuracy

print('le score est de : ',model.score(X,y))

#Test du modele
imnum=1001
test = digits['images'][imnum].reshape(1, -1)
print('le chiffre est un :')
plt.figure()
plt.imshow(digits['images'][imnum], cmap = 'Greys_r')
plt.show()

print('la prédiction est un : ',model.predict(test))
```
